```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
colors = [
    '#065e60', '#0a9396', '#99d5c9', '#ee9b00', '#ffb833', '#99d6ff', '#0093f5', '#00568f'
]

# Set general font size
SMALL_SIZE = 14
MEDIUM_SIZE = 16
BIGGER_SIZE = 22

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=MEDIUM_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

f, ax = plt.subplots(1, 1, figsize=(4, 4))

indices = np.arange(len(colors))

for i, c in enumerate(colors):
    ax.bar(indices[i], 1, color=c)

ax.set_xticks(indices)
plt.show()
```

```python
# relu function
def relu(x):
    return np.maximum(0,x)

# sigmoid function
def sigmoid(x):
    return 1/(1+np.exp(-x))

# tanh function
def tanh(x):
    return np.tanh(x)

x = np.arange(-5,5.1,0.1)

alpha = 0.3

f, ax = plt.subplots(1, 3, figsize=(12, 3.5))

ax[0].plot(x, relu(x), color=colors[1])
ax[1].plot(x, sigmoid(x), color=colors[4])
ax[2].plot(x, tanh(x), color=colors[6])

ax[0].set_title('ReLU')
ax[1].set_title('Sigmoid')
ax[2].set_title('Tanh')

ax[0].set_xlim(-5, 5)
ax[1].set_xlim(-5, 5)
ax[2].set_xlim(-5, 5)

ax[0].set_ylim(ax[0].get_ylim())
ax[1].set_ylim(ax[1].get_ylim())
ax[2].set_ylim(ax[2].get_ylim())

ax[0].hlines(0, -5, 5, linestyles='dashed', colors='k', alpha=alpha, zorder=-10)
ax[1].hlines(0, -5, 5, linestyles='dashed', colors='k', alpha=alpha, zorder=-10)
ax[2].hlines(0, -5, 5, linestyles='dashed', colors='k', alpha=alpha, zorder=-10)

ax[0].vlines(0, -5, 6, linestyles='dashed', colors='k', alpha=alpha, zorder=-10)
ax[1].vlines(0, -5, 6, linestyles='dashed', colors='k', alpha=alpha, zorder=-10)
ax[2].vlines(0, -5, 6, linestyles='dashed', colors='k', alpha=alpha, zorder=-10)

plt.tight_layout()
plt.savefig('random_images/activation_functions.png', dpi=300)

plt.show()
```
